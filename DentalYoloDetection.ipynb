```python
# Instalează bibliotecile necesare
!pip install ultralytics gradio

# Importă YOLOv8 și gradio
from ultralytics import YOLO
import gradio as gr
from PIL import Image

# Încarcă modelul YOLO
model = YOLO('/content/best.pt')
# Funcția de predicție care primește o imagine și returnează imaginea cu predicții
def detect_objects(image):
    results = model(image, conf=0.5)  # Ajustează confidența dacă este necesar
    result_image = results[0].plot()  # Obține imaginea cu predicțiile
    return Image.fromarray(result_image)

# Creează interfața Gradio
interface = gr.Interface(
    fn=detect_objects,
    inputs=gr.Image(type="pil"),
    outputs="image",
    title="Dental Analysis YOLOv8 Detector",
    description="Încarcă o imagine pentru a detecta elementele cu modelul YOLOv8."
)

# Lansează aplicația
interface.launch()
```
