```python
# InstaleazÄƒ bibliotecile necesare
!pip install ultralytics gradio

# ImportÄƒ YOLOv8 È™i gradio
from ultralytics import YOLO
import gradio as gr
from PIL import Image

# ÃŽncarcÄƒ modelul YOLO
model = YOLO('/content/best.pt')
# FuncÈ›ia de predicÈ›ie care primeÈ™te o imagine È™i returneazÄƒ imaginea cu predicÈ›ii
def detect_objects(image):
    results = model(image, conf=0.5)  # AjusteazÄƒ confidenÈ›a dacÄƒ este necesar
    result_image = results[0].plot()  # ObÈ›ine imaginea cu predicÈ›iile
    return Image.fromarray(result_image)

# CreeazÄƒ interfaÈ›a Gradio
interface = gr.Interface(
    fn=detect_objects,
    inputs=gr.Image(type="pil"),
    outputs="image",
    title="Dental Analysis YOLOv8 Detector",
    description="ÃŽncarcÄƒ o imagine pentru a detecta elementele cu modelul YOLOv8."
)

# LanseazÄƒ aplicaÈ›ia
interface.launch()
```
